```python
def multicxcpcheck(dong,multidict,): #다중 동 먼저 체크해서 해당될경우 처리하고 진행
    if dong in multidict:
        for plus in multidict[dong]:
            countdict[plus] += (1/len(multidict[dong]))
        return True
    return False
    
def singlecheck(dong,singledict):
    singlekey=list(singledict.keys())
    for sdong in singlekey:
        for slist in singledict[sdong]:
            if str(dong) in slist:
                countdict[sdong] +=1
                return True
    return False

#이하 2개 함수는 속도가 느릴 수 있음
def usefindmulti(dong,multidict):
    for bub in list(multidict.keys()):
        if str(dong).find(bub) != -1:      
            for plus in multidict[dong]:
                countdict[plus] += (1/len(multidict[dong]))
        return True
    return False

def usefindsingle(dong,singledict):
    singlekey=list(singledict.keys())
    for sdong in singlekey:
        for slist in singledict[sdong]:
            for bubdong in slist:
                if str(dong).find(bubdong) != -1:
                    countdict[sdong] +=1
                    return True
    return False
##
def myeoncheck(myeonname,myeonlist):
    if str(myeonname) in myeonlist:
        countdict[myeonname]+=1
        return True
    return False

def myeonfind(myeonname,myeonlist):
    for word in myeonlist:
        if myeonname.find(word) !=-1:
            countdict[word]+=1
            return True
    return False
   
def makecountlist(datalist):#법정동 이름 리스트를 변수로
    errlist=[]
    for data in datalist:
         if multicxcpcheck(data,multi):
                continue
         elif singlecheck(data,single):
                continue
         elif myeoncheck(data,myeon):
                continue
         else:
            errlist.append(data)
    return countdict,errlist
            
def makecountlist2(datalist):#보류
    errlist=[]
    for data in datalist:
        if multicxcpcheck(data,multi):
                continue
        elif singlecheck(data,single):
                continue
        elif myeoncheck(data,myeon):
                continue
        elif usefindmulti(data,multi):
                continue
        elif usefindsingle(data,single):
                continue
        elif myeonfind(data,myeon):
                continue
        else:
            errlist.append(data)
    return countdict,errlist
            
            
            
    
```
